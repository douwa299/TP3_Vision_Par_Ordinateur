```python
import cv2

cap = cv2.VideoCapture("testVideo.mp4")  # 0 = webcam, or replace with "video.mp4"
if not cap.isOpened():
    print("❌ Webcam non disponible")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('Webcam (Grayscale)', gray)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

Que se passe-t-il si vous utilisez une vidéo enregistrée à la place de la webcam ?

Le même code fonctionne, seul le paramètre de VideoCapture() change

Webcam : VideoCapture(0) (flux temps réel)

Vidéo : VideoCapture("chemin/video.mp4") (lecture fichier)

La vidéo s'arrête automatiquement à la fin du fichier

waitKey(25) au lieu de waitKey(1) pour respecter le FPS original
