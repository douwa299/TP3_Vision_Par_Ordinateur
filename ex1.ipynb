```python
import cv2
import numpy as np

print("=== EXERCICE 1: Analyse d'une image ===")

# Create a test image with various shapes for better edge detection
img = np.ones((500, 700, 3), dtype=np.uint8) * 255  # White background

# Draw different shapes
cv2.rectangle(img, (50, 50), (200, 200), (255, 0, 0), -1)    # Blue rectangle
cv2.circle(img, (400, 150), 80, (0, 255, 0), -1)            # Green circle  
cv2.ellipse(img, (200, 350), (100, 50), 45, 0, 360, (0, 0, 255), -1)  # Red ellipse
cv2.putText(img, 'OpenCV Test', (350, 400), cv2.FONT_HERSHEY_SIMPLEX, 1, (128, 0, 128), 2)

print(" Image de test creee")

# 1. Afficher l'image originale
cv2.imshow('1. Image originale', img)
print(" Image originale - Appuyez sur une touche ou attendez 5s")
key = cv2.waitKey(5000)  # 5 second timeout
if key == -1:
    print("  Timeout - passage a l'etape suivante")

# 2. Conversion en niveaux de gris
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('2. Image en niveaux de gris', gray)
print(" Image niveaux de gris - Appuyez sur une touche ou attendez 5s")
key = cv2.waitKey(5000)
if key == -1:
    print(" Timeout - passage a l'etape suivante")

# 3. Detection des contours avec Canny
edges = cv2.Canny(gray, 100, 200)
cv2.imshow('3. Contours detectes (Canny)', edges)
print("Contours Canny - Appuyez sur une touche ou attendez 5s")
key = cv2.waitKey(5000)
if key == -1:
    print(" Timeout - fermeture des fenetres")

# Fermer toutes les fenetres
cv2.destroyAllWindows()

# Sauvegarder les resultats
cv2.imwrite('image_originale.jpg', img)
cv2.imwrite('image_grise.jpg', gray)
cv2.imwrite('contours_canny.jpg', edges)

print("\n EXERCICE 1 TERMINE!")
print("Fichiers sauvegardes:")
print("   - image_originale.jpg")
print("   - image_grise.jpg") 
print("   - contours_canny.jpg")
```
